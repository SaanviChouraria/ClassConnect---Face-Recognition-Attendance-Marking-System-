```python
import cv2
import face_recognition
import os
import numpy as np
from datetime import datetime

path = r'C:\Users\Saanvi\Desktop\Coding Classes\AI Advance Internship\Projects - Class Attendance System\Training Images'

# Initialize empty lists to store images and class names
images = []
classNames = []

# Get the list of files in the specified directory
mylist = os.listdir(path)

# Iterate through each file in the directory
for cl in mylist:

    # Read the image using OpenCV (cv2)
    curImg = cv2.imread(f'{path}/{cl}')

    # Append the image to the 'images' list
    images.append(curImg)

    # Extract the class name from the filename (remove extension)
    classNames.append(os.path.splitext(cl)[0])


# Defines a function named findEncodings that takes a list of images (images) as its parameter
def findEncodings(images):

    # Initializes an empty list called encodeList to store the face encodings
    encodeList = []

    # Initiates a loop that iterates over each image in the images list
    for img in images:

        # Converts the image from BGR (OpenCV's default color format) to RGB
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        # Uses the face_recognition library to find the face encoding for the given image.
        # The [0] indexing assumes that there is only one face in each image.
        # The resulting face encoding is stored in the variable encoded_face
        encoded_face = face_recognition.face_encodings(img)[0]

        # Appends the face encoding (encoded_face) to the encodeList
        encodeList.append(encoded_face)

    # Exits the function and returns the list of face encodings.
    return encodeList


# Calls the findEncodings function with the images list as an argument and stores the resulting list
# of face encodings in the variable encoded_face_train
encoded_face_train = findEncodings(images)


# Defines a function named markAttendance that takes a name as its parameter
def markAttendance(name):
    
    # Opens the CSV file named 'Attendance.csv' in append mode ('a+')
    with open(r"C:\Users\Saanvi\Desktop\Coding Classes\AI Advance Internship\Projects - Class Attendance System\Attendance.csv",'a+') as f:
        
        # Gets the current date and time using the datetime module.
        now = datetime.now()
        
        # Formats the current time as 'hour:minute:second:AM/PM'
        time = now.strftime('%I:%M:%S:%p')
        
        # Formats the current date as 'day-month-year'
        date = now.strftime('%d-%B-%Y')
        
        # Writes a new line to the CSV file with the provided name, time, and date
        f.write(f'\n{name}, {time}, {date}')


# take pictures from webcam
cap = cv2.VideoCapture(0)

# Infinite loop for continuously capturing frames
while True:
    
    # Read a frame from the webcam
    ret, img = cap.read()
    if not ret:
        break

    # Resize the frame to speed up face recognition
    imgS = cv2.resize(img, (0,0), None, 0.25, 0.25)

    # Convert the resized frame to RGB format (required by face_recognition library)
    imgS = cv2.cvtColor(imgS, cv2.COLOR_BGR2RGB)

    # Detect faces in the resized frame
    faces_in_frame = face_recognition.face_locations(imgS)
   
    # Encode the faces in the frame
    encoded_faces = face_recognition.face_encodings(imgS, faces_in_frame)

    # Define a threshold for confidence level
    threshold = 0.5

    # Iterate through each detected face and its corresponding encoding
    for encode_face, faceloc in zip(encoded_faces, faces_in_frame):

        # Compare the detected face with the known face encodings    
        matches = face_recognition.compare_faces(encoded_face_train, encode_face)
        faceDist = face_recognition.face_distance(encoded_face_train, encode_face)

        # If there is a match and the confidence level is above the threshold
        if any(matches) and min(faceDist) < threshold:
            matchIndex = np.argmin(faceDist)
            name = classNames[matchIndex].upper().lower()
        else:
            name = "Image not found"

        y1, x2, y2, x1 = faceloc

        # since we scaled down by 4 times
        y1, x2, y2, x1 = y1 * 4, x2 * 4, y2 * 4, x1 * 4
        cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)
        cv2.rectangle(img, (x1, y2-35), (x2, y2), (0, 255, 0), cv2.FILLED)
        cv2.putText(img, name, (x1+6, y2-5), cv2.FONT_HERSHEY_COMPLEX, 1, (255, 255, 255), 2)
        markAttendance(name)
    
    cv2.imshow('webcam', img)

    # Break the loop if 'q' key is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the VideoCapture and close all OpenCV windows
cap.release()
cv2.destroyAllWindows()
```

```python
import tkinter as tk
from tkinter import ttk
from PIL import Image, ImageTk
import cv2
import face_recognition
import os
import numpy as np
from datetime import datetime
import csv
from tkinter import filedialog

path = r"C:\Users\Saanvi\Desktop\Coding Classes\AI Advance Internship\Projects - Class Attendance System\Training Images"
images = []
classNames = []
mylist = os.listdir(path)
for cl in mylist:
    curImg = cv2.imread(f'{path}/{cl}')
    images.append(curImg)
    classNames.append(os.path.splitext(cl)[0])

def findEncodings(images):
    encodeList = []
    for img in images:
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        encoded_face = face_recognition.face_encodings(img)[0]
        encodeList.append(encoded_face)
        
    return encodeList

encoded_face_train = findEncodings(images)

root = tk.Tk()
root.title("ClassConnect")

canvas_width = 1000
canvas_height = 750

homepage_path = r"C:\Users\Saanvi\Desktop\Coding Classes\AI Advance Internship\Projects - Class Attendance System\HomeScreen.png"
homepage_img = Image.open(homepage_path)
homepage_img = homepage_img.resize((canvas_width, canvas_height))
homepage_img = ImageTk.PhotoImage(homepage_img)

canvas1 = tk.Canvas(root, width=canvas_width, height=canvas_height)
canvas1.pack()  
canvas1.create_image(0, 0, anchor=tk.NW, image=homepage_img)

class_connect_2_frame = tk.Frame(root, width=canvas_width, height=canvas_height, bg='light blue')

attendance_records = {}

def update_attendance_display():
    for i in tree.get_children():
        tree.delete(i)
    for student, details in attendance_records.items():
        tree.insert("", "end", values=(student, details['Date'], details['Time']))

def switch_to_class_connect_2():
    canvas1.pack_forget()
    class_connect_2_frame.pack()
    label1 = tk.Label(class_connect_2_frame, text='Welcome to ClassConnect. Please click on the button below to start taking the attendance of your class.',
                      font=('Arial Rounded MT Bold', 14), bg='yellow')
    label1.place(x=20, y=30)

def update_attendance_csv():
    with open(r"C:\Users\Saanvi\Desktop\Coding Classes\AI Advance Internship\Projects - Class Attendance System\Attendance.csv", mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(['Name', 'Date', 'Time'])
        for student, details in attendance_records.items():
            writer.writerow([student, details['Date'], details['Time']])

def take_attendance():
    cap = cv2.VideoCapture(0)
    date = datetime.now().strftime("%d-%B-%Y")
    time = datetime.now().strftime("%I:%M:%S %p")
    while True:
        ret, img = cap.read()
        if not ret:
            break
        imgS = cv2.resize(img, (0, 0), None, 0.25, 0.25)
        imgS = cv2.cvtColor(imgS, cv2.COLOR_BGR2RGB)
        faces_in_frame = face_recognition.face_locations(imgS)
        encoded_faces = face_recognition.face_encodings(imgS, faces_in_frame)
        threshold = 0.4
        names_in_frame = []
        for encode_face, faceloc in zip(encoded_faces, faces_in_frame):
            matches = face_recognition.compare_faces(encoded_face_train, encode_face)
            faceDist = face_recognition.face_distance(encoded_face_train, encode_face)
            if any(matches) and min(faceDist) < threshold:
                matchIndex = np.argmin(faceDist)
                name = classNames[matchIndex].upper().lower()
            else:
                name = "Unknown"
            names_in_frame.append(name)
            # Draw rectangle around the face
            y1, x2, y2, x1 = faceloc
            y1, x2, y2, x1 = y1 * 4, x2 * 4, y2 * 4, x1 * 4
            cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)
            cv2.rectangle(img, (x1, y2 - 35), (x2, y2), (0, 255, 0), cv2.FILLED)
            cv2.putText(img, name, (x1 + 6, y2 - 5), cv2.FONT_HERSHEY_COMPLEX, 1, (255, 255, 255), 2)
            for name in names_in_frame:
                attendance_records[name] = {'Date': date, 'Time': time}
        update_attendance_display()
        # Display the captured frame with OpenCV
        cv2.imshow('Attendance', img)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    cap.release()
    cv2.destroyAllWindows()
    update_attendance_csv()


def register_new_student():
    register_window = tk.Toplevel(root)
    register_window.title("Register New Student")
    def register_student():
        student_name = entry.get()
        cap = cv2.VideoCapture(0)
        while True:
            ret, img = cap.read()
            if not ret:
                break
            cv2.imshow('Register New Student', img)
            if cv2.waitKey(1) & 0xFF == ord('q'):
                filename = f"{student_name}.jpg"
                image_path = os.path.join(path, filename)
                cv2.imwrite(image_path, img)
                classNames.append(student_name)
                img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
                encoded_face = face_recognition.face_encodings(img)[0]
                encoded_face_train.append(encoded_face)
                break
        cap.release()
        cv2.destroyAllWindows()
        register_window.destroy()
        
    label = tk.Label(register_window, text="Enter the name of the new student:")
    label.pack()
    entry = tk.Entry(register_window)
    entry.pack()
    register_button = tk.Button(register_window, text="Register", command=register_student)
    register_button.pack()



def download_attendance():
    # Function to handle downloading attendance as a CSV file
    update_attendance_csv()
   
    # Choose the file location for saving the CSV file
    file_path = filedialog.asksaveasfilename(defaultextension=".csv", filetypes=[("CSV files", "*.csv")])
   
    if file_path:
        # If a file location is selected, write the attendance data to the CSV file
        with open(file_path, mode='w', newline='') as file:
            writer = csv.writer(file)
            writer.writerow(['Name', 'Date', 'Time'])
            for student, details in attendance_records.items():
                writer.writerow([student, details['Date'], details['Time']])
    else:
        print("No file selected.")
       
def show_help():
    # Function to display instructions on how to use the app
    help_window = tk.Toplevel(root)
    help_window.title("Help")

    help_text = """Welcome to ClassConnect!\n\nHere's how to use the app:\n\n
    1. Register New Student: Click this button to add a new student to the system.
    You will be prompted to enter the student's name, and then you can take a picture of the student's face for recognition.

    2. Take Attendance: Click this button to start taking attendance. The app will use facial recognition to identify students present in the class.

    3. File: Click this button to download the attendance records displayed in the table as a CSV file.

    4. Exit: Click this button to close the application.\n\n
    Have a good day!\n"""

    help_label = tk.Label(help_window, text=help_text, justify='left')
    help_label.pack()


register_new_student_button = tk.Button(class_connect_2_frame, text='Register New Student', bg='light green', font=('Bahnschrift', 16), command=register_new_student)
register_new_student_button.place(x=400, y=200)

take_attendance_button = tk.Button(class_connect_2_frame, text='Take Attendance', bg='pink', font=('Bahnschrift', 16), command=take_attendance)
take_attendance_button.place(x=425, y=120)


file_button = tk.Button(class_connect_2_frame, text='File', bg='orange', font=('Bahnschrift', 13), command=download_attendance)
file_button.place(x=30, y=80)

help_button = tk.Button(class_connect_2_frame, text='Help', bg='orange', font=('Bahnschrift', 13), command=show_help)
help_button.place(x=90, y=80)

tree = ttk.Treeview(class_connect_2_frame, columns=('Name', 'Date', 'Time'), show='headings')
tree.heading('Name', text='Name', anchor='center')
tree.heading('Date', text='Date')
tree.heading('Time', text='Time')
tree.place(x=50, y=300, width=800, height=350)

exit_button = tk.Button(class_connect_2_frame, text="Exit", bg='red', font=('Bahnschrift', 16), command=root.destroy)
exit_button.place(x=450, y=680)

root.after(5000, switch_to_class_connect_2)

root.mainloop()
```
